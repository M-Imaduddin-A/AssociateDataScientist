```python
dir = 'hungarian.data'
```

```python
with open(dir, encoding= 'Latin1') as file:
   lines = [line.strip() for line in file]
lines[0:10]
```

```python
data = itertools.takewhile(
  lambda x :len(x)==76,
 (' '.join(lines[i:(i +10)]).split() for i in range(0,len(lines),   10)))
df = pd.DataFrame.from_records(data)
df.head()
```

```python
# @title Menampilkan informasi data
df.info()
```

```python
# @title penghapusan data set
df = df.iloc[:,:-1]
df = df.drop(df.columns[0],axis=1)
```

```python
# @title Mengubah ke float
df = df.astype(float)
df.info()
```
